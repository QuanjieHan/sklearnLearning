```python
X_train=[[6],[8],[10],[14],[18]]
y_train=[[7],[9],[13],[17.5],[18]]

from sklearn.linear_model import LinearRegression
linearRegressor=LinearRegression()
print(linearRegressor)

linearRegressor.fit(X_train,y_train)

import numpy as np

xx=np.linspace(0,26,100)
# print(xx.shape)
xx=xx.reshape(xx.shape[0],1)
# help(np.reshape)
# print(xx.shape)
yy=linearRegressor.predict(xx)

import matplotlib.pyplot as plt
plt.scatter(X_train,y_train)
plt1,=plt.plot(xx,yy,label='Degree1')
plt.axis([0,25,0,25])
plt.xlabel('Diameter of Pizza')
plt.ylabel('Price of Pizza')
plt.legend(handles=[plt1])
plt.show()

print("The R-Squared value of Linear Regression performing on the training data is: ", linearRegressor.score(X_train,y_train))
```

```python
# using order two polynomial
from sklearn.preprocessing import PolynomialFeatures
poly2=PolynomialFeatures(degree=2)
# help(PolynomialFeatures)
print(poly2)
X_train_poly2=poly2.fit_transform(X_train)

regressor_poly2=LinearRegression()
regressor_poly2.fit(X_train_poly2,y_train)
xx_poly2=poly2.transform(xx)
yy_poly2=regressor_poly2.predict(xx_poly2)

plt.scatter(X_train,y_train)
plt1,=plt.plot(xx,yy,label='Degree1')
plt2,=plt.plot(xx,yy_poly2,label='Degree2')
plt.axis([0,25,0,25])
plt.xlabel('Diameter of Pizza')
plt.ylabel('Price of Pizza')
plt.legend(handles=[plt1,plt2])
plt.show()

print("The R-Squared value of polynomial Regressor (Degree 2) performing on the training data is :", regressor_poly2.score(X_train_poly2,y_train))
```

```python
# 4th-order polynomial regression
poly4=PolynomialFeatures(degree=4)
regressor_poly4=LinearRegression()

X_train_poly4=poly4.fit_transform(X_train)
regressor_poly4.fit(X_train_poly4,y_train)

xx_poly4=poly4.transform(xx)
yy_poly4=regressor_poly4.predict(xx_poly4)

plt.scatter(X_train,y_train)
plt1,=plt.plot(xx,yy,label='degree1')
print(plt1)
plt2,=plt.plot(xx,yy_poly2,label='degree2')
plt3,=plt.plot(xx,yy_poly4,label='degree4')
print(plt3)
#help(plt.plot)
plt.axis([0,25,0,25])
plt.xlabel('Diameter of Pizza')
plt.ylabel('Price of  Pizza')
plt.legend(handles=[plt1,plt2,plt3])
plt.show()

print("The R-Squared value of polynomial Regressor (Degree 4 ) performing on the training data is :", regressor_poly4.score(X_train_poly4,y_train))
```

```python
# Performing on the test data set
X_test=[[6],[8],[11],[16]]
y_test=[[8],[12],[15],[18]]

print("The R_Squared value of Linear Regressor performing on test data set is :",linearRegressor.score(X_test,y_test))
```

```python
X_test_poly2=poly2.transform(X_test)
print("The R-Squared value of Polynomial Regression (Degree 2) performing on the test data set is :", regressor_poly2.score(X_test_poly2,y_test))
```

```python
X_test_poly4=poly4.transform(X_test)
print("The R-Squared value of Polynomial Regression (Degree 4) performing on the test data set is: ", regressor_poly4.score(X_test_poly4,y_test))
```

```python
# Lasso regression
from sklearn.linear_model import Lasso
lasso_poly4=Lasso()
print(lasso_poly4)
lasso_poly4.fit(X_train_poly4,y_train)
print("The R-Squared value of Lasso regression of Degree 4 performing on the training data is: ", lasso_poly4.score(X_train_poly4,y_train))
print("The R-Squared value of Lasso regression of Degree 4 performing on the test data is :", lasso_poly4.score(X_test_poly4,y_test))
```

```python
print(lasso_poly4.coef_)
```

```python
print(regressor_poly4.coef_)
```

```python
#  Ridge Regression
print(regressor_poly4.coef_)
print(np.sum(regressor_poly4.coef_**2))
```

```python
from sklearn.linear_model import Ridge
ridge_poly4=Ridge()
print(ridge_poly4)

ridge_poly4.fit(X_train_poly4,y_train)
print("The R-Squared value of Ridge Regression of Degree 4 performing on the training data is :", ridge_poly4.score(X_train_poly4,y_train))
print("The R-Squared value of Ridge Regression of Degree 4 performing on the test data is :", ridge_poly4.score(X_test_poly4,y_test))
```
